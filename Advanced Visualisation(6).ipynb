```python
movie.columns=['Film','Genre','CriticRating','AudienceRatings','BudgetMillion','Year']
movie.head()
```

```python
movie.Genre =movie.Genre.astype('category')
movie.BudgetMillion =movie.BudgetMillion.astype('long')
movie.Year =movie.Year.astype('category')
```

```python
#back to the scatter plot
plot18=sb.FacetGrid(movie, row='Genre',col='Year',hue='Genre')
kws=dict(s=50,linewidth=0.5, edgecolor='black')
plot18=plot18.map(plt.scatter,'CriticRating','AudienceRatings',**kws)
plot18.set(xlim=(0,100),ylim=(0,100))
# here we are generating the keyword(i.e kws) which contains the different datas like size edge color ect.
#then passing it in the map parameter


for ax in plot18.axes.flat:
    ax.plot((0,100), (0,100), c='black', ls='--')
plot18.add_legend()
```
