2-Write a python program that identifies the last Thursday of the month. Your program must include two functions:
   isLastThursday(input_date) -
           -This function must return True if the given date is the last Thursday of the respective month.
           -Input_date is of type String.

   getLastThursday(input_month) -
           -This function must return the last Thursday of the respective month.
           -Input_month is of type String. Function must return a date in string format.
Note : Refer datetime library to make conversions between String and Date formats.



```python

import datetime
import calendar




def isLastThursday(input_date):
    date_time_obj1 = datetime.datetime.strptime(input_date, '%d/%m/%y')
    last_thursday = max(week[3] for week in calendar.monthcalendar(date_time_obj1.year, date_time_obj1.month))
    check_str = str(last_thursday) + '/' + str(date_time_obj1.month) + '/' + str(date_time_obj1.year)
    
    if input_date==(check_str):
        return True
    else:
        return False


def getLastThursday(input_month):
    date_time_obj2 = datetime.datetime.strptime(input_month, '%b')
    last_thursday = max(week[3] for week in calendar.monthcalendar(2021, date_time_obj2.month))
    check_str = str(last_thursday) + '/' + str(date_time_obj2.month) + '/' + str(2021)
    return check_str


print(isLastThursday('26/05/21'))
print(getLastThursday('Mar'))
```
